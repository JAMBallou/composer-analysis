```python
# --- 01_data_exploration.ipynb ---
# Explore MAESTRO dataset metadata and sample recordings.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import librosa.display
import pretty_midi
from IPython.display import Audio

# Load MAESTRO metadata
metadata_path = "../data/maestro-v3.0.0/maestro-v3.0.0.csv"
df = pd.read_csv(metadata_path)

# --- Basic overview ---
df.head()
df.info()
df.describe()

# --- Composer distribution ---
composer_counts = df["composer"].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=composer_counts.index, y=composer_counts.values)
plt.xticks(rotation=45)
plt.title("Composer Distribution in MAESTRO Dataset")
plt.show()

# --- Listen to a sample audio ---
sample_path = df.iloc[0]["audio_filename"]
# TODO: adjust path if relative to your system
y, sr = librosa.load(f"../data/maestro-v3.0.0/{sample_path}", sr=22050)
Audio(y, rate=sr)
```
